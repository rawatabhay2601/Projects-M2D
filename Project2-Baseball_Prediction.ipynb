```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 
import sklearn
import scipy 
from sklearn.model_selection import train_test_split 
import warnings
warnings.filterwarnings('ignore')
```

```python
df_bas= pd.read_csv('baseball.csv')
df_bas
```

```python
import missingno as msno

msno.bar(df_bas)
```

```python
#Since no null values available we can move to checking the properties and relations in the datasets
df_bas.describe(include='all')
```

```python
#Checking relations between wins and runs
def bar_relations(x1,y1):
    plt.figure(figsize=(8,8))
    sns.barplot(x=x1,y=y1,data=df_bas)
    return plt.show()


for col in df_bas.columns:
    if col != 'W':
        print(col,' vs W\n')
        bar_relations(col,'W')
```

```python
#Doubt : What does the line on bar denotes
#Above graphs show relation of wins with every column
#Now let's remove the outliers 
from scipy.stats import zscore
z = np.abs(zscore(df_bas))
z
```

```python
threshold = 3
np.where(z>3)
```

```python
z[5][1]
```

```python
df_new = df_bas[(z<3).all(axis=1)]
df_new
```

```python
#outliers are removed so now we will check for skewness
df_new.skew()
```

```python
#Removing skewness
from sklearn.preprocessing import PowerTransformer
pt = PowerTransformer()

for col in df_new.columns:
    if df_new.loc[:,col].skew() > 0.55:
        df_new[col] = pt.fit_transform(df_new[col].to_frame())

df_new.skew()
```

```python
#Hence skewness is removed and now we'll transform the data
#But first split x and y variable inputs
x = df_new.iloc[:,1:-1]
x
```

```python
y = df_new.loc[:,'W']
y
```

```python
#Now we shall transform the data
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
df_x = sc.fit_transform(x)
df_x = pd.DataFrame(df_x , columns=x.columns)
df_x
```

```python
#Again checking skewness after transforming the data
df_x.skew()
```

```python
#Now we shall move forward with testing of models 
#Since the output variable contains continous values we shall use regression techniques
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.ensemble import AdaBoostRegressor
```

```python
x_train,x_test,y_train,y_test = train_test_split(df_x, y, test_size=0.2, random_state=42)
```

```python
models = [LinearRegression(),DecisionTreeRegressor(),KNeighborsRegressor(),RandomForestRegressor()]
for m in models:
    m.fit(x_train,y_train)
    pred = m.predict(x_test)
    r2 = r2_score(y_test,pred)
    mae = mean_absolute_error(y_test,pred)
    mse = mean_squared_error(y_test,pred)
    rmse = np.sqrt(mse)
    print(m,'\n r2 score : ',r2,'\n mean abs err : ',mae,'\n mean squared err : ',mse,'\nroot mean squared error : ',rmse,'\n\n\n')
```

```python
#From above we can see that linear regression is giving the best output
#Now we can look for the best random state
random = 0
r2score = 0
for i in range(0,200):
    x_train,x_test,y_train,y_test = train_test_split(df_x, y, test_size=0.2, random_state=i)
    lm = LinearRegression()
    lm.fit(x_train,y_train)
    pred = lm.predict(x_test)
    r2 = r2_score(y_test,pred)
    if r2 > r2score:
        r2score = r2
        random = i
print('Best random state',random,'for best r2 score',r2score)
```

```python
#Now let's check for cross val score
lm = LinearRegression()
cvs = cross_val_score(lm,df_x,y,cv = 5,scoring='r2').mean()
print(cvs)
```

```python
#We can see that linear regression is overfitting the model so we'll try to boost the result by-
#1 -- Finding best parameters 
#2 -- Using regularization technique to remove overfitting
def best_parameter(model,para):
    from sklearn.model_selection import GridSearchCV
    
    grd = GridSearchCV(model,para,cv=5,scoring='r2')
    grd.fit(df_x,y)
    return [grd.best_params_, grd.best_score_]
```
